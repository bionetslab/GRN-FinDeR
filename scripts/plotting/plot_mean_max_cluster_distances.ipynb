```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
max_df = pd.read_csv("../../results/manuscript/cluster_diameters_all_tissues.tsv", sep='\t', index_col=0)
plt.figure(figsize=(7, 4))
ax = sns.boxplot(data=max_df, x=max_df["num_clusters"].values, y=max_df["max_intra_dist"].values, legend=False, showfliers=False, boxprops=dict(facecolor='lightgrey', alpha=0.6, edgecolor='grey'), width=0.4)
ax.set_xticklabels([1,'', '', '', 5, '', '', '', '', 10, '', '', '', 50, '', '', '', '', 100])
ax.set_xlabel("Number of target gene clusters", fontsize=19)
ax.set_ylabel("Max. cluster distance", fontsize=19)
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
plt.tight_layout()
plt.savefig("max_cluster_distances.pdf")
```

```python
mean_df = pd.read_csv("../../results/manuscript/mean_cluster_distances_all_tissues.tsv", sep='\t', index_col=0)
plt.figure(figsize=(7, 4))
ax = sns.boxplot(data=mean_df, x="num_clusters", y="mean_intra_dist", legend=False, showfliers=False, boxprops=dict(facecolor='lightgrey', alpha=0.6, edgecolor='grey'), width=0.4)
ax.set_xticklabels([1,'', '', '', 5, '', '', '', '', 10, '', '', '', 50, '', '', '', '', 100])
ax.set_xlabel("Number of target gene clusters", fontsize=19)
ax.set_ylabel("Mean cluster distance", fontsize=19)
ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)
plt.tight_layout()
plt.savefig("mean_cluster_distances.pdf")
```
